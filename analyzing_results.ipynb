We will verify from the tagged Etherscan names which are correct and which are incorrect

```python
remove_words = ["token",
"proxy",
"contract",
"work",
"protocol",
"governance",
"dao",
"finance"]

```

```python
def to_lower(words):
    lower_words = []
    for word in words:
        lower_words.append(word.lower())
    return lower_words

def filter_key_words(p_name):
    project_name = p_name.lower()
    project_name = project_name.split(' ')
    for name in project_name:
        if name in remove_words:
            project_name.remove(name)
    return project_name

def verify_name(p_name, labels,names):
    filtered_p_name = filter_key_words(p_name)
    labels = to_lower(labels)
    # names = to_lower(names)
    # labels = labels.lower()
    names = names.lower()
    is_labeled = None
    is_name = None
    for name in filtered_p_name:
        if name in labels:
            is_labeled = True
        if name in names:
            is_name = True
    if is_labeled and is_name:
        return "both"
    elif is_labeled ==True:
        return "in_label"
    elif is_name == True:
        return "in_name"
    else:
        return "not_tagged"
```

```python
df_is_tagged['verified'] = df_is_tagged.apply(lambda row: verify_name(row['project_name'],row['labels'],row['name']),axis=1)
df_is_tagged
```

```python
stats = df_is_tagged.groupby('verified')['other_created_address'].count()
df_verified_stats = pd.DataFrame(stats)
df_verified_stats['percentage'] = df_is_tagged.groupby('verified')['other_created_address'].count()/df_is_tagged['other_created_address'].count()
```
